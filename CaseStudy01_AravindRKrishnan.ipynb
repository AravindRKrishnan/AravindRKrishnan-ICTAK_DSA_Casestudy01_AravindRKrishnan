# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
sns.barplot(x=data['Buyer Gender'].value_counts().index,y=data['Buyer Gender'].value_counts().values)
plt.title('Male vs Female Buyers participated in the sales')
plt.ylabel('Number')
plt.xlabel('Gender')
plt.legend(loc=1)
plt.show()
```
